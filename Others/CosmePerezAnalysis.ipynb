```python

html_value = """
<div id="anyo-Introducción"><b>Introducción</b><br> </div>Aunque su verdadero nombre era Cosme Pérez, fue conocido artísticamente como 'Juan Rana', según documenta ya la <em><strong>Genealogía</strong></em> (G, 117) —este apodo ya se le aplicaba en 1636—. Aunque Cotarelo aventuró que pudo haber nacido en Madrid (CM5, CLVIII), hoy sabemos que nació en Tudela de Duero (Valladolid), en cuya iglesia parroquial fue bautizado el 7 de abril de 1593. Fue hijo de Damián Pérez e Isabel de Basto, su segunda esposa, y fue el cuarto de una familia de cuatro hermanos (F. Sáez Raposo, <em><strong>Juan Rana y el teatro cómico breve del siglo XVII</strong></em>, Madrid, Fundación Universitaria Española, 2005, 21-23).
<br>Según algunas fuentes, como Sánchez Arjona y Subirá, estuvo casado con una actriz llamada Bernarda Ramírez (SA, 330; Su, 78). Por su lado, Rennert puso en duda su matrimonio con Bernarda Ramírez y creyó más probable que estuviese casado con Bernarda Manuela ['la Grifona'], que sería, según esto, su primera mujer (R, 554). Sin embargo, ambos matrimonios no se encuentran documentados, como ya puso de relieve Cotarelo (CM5, CLX), y probablemente se fundan en deducciones realizadas a partir de algunos entremeses en los que Cosme Pérez figura junto con estas actrices haciendo papeles de marido y mujer. De hecho, la <em><strong>Genealogía</strong></em> tan sólo documenta su matrimonio con María de Acosta, con la que tuvo una hija —y quizá también un hijo: véase el año 1634—, Francisca María Pérez, y se refiere a la circunstancia de que los tres fueron recibidos en la Cofradía de Nuestra Señora de la Novena estando en la compañía de Tomás Fernández [Cabredo] (G, 117, 377, 408-09) —la fecha en que fueron recibidos como cofrades no se especifica en la <em><strong>Genealogía</strong></em>, aunque Cotarelo la sitúa en 1631 (CM5, CLVIII)—.
<br>En la misma <em><strong>Genealogía</strong></em> se hace alusión a su enorme popularidad en los siguientes términos: "Fue mui zelebrado en la parte de grazioso, y aun excedió a todos los de su tiempo, y solo con salir a las tablas, y sin hablar, probocaba a risa y al aplauso a los que le veían" (G, 117). Como muestra de su popularidad, el autor de la <em><strong>Genealogía</strong></em> nos cuenta que el actor Domingo Canejil, que murió en 1707, recibió el nombre de 'Ranilla' "porque decían que ni Juan Rana había llegado a su grande habilidad" (G, 171). Según la misma fuente, gozó de la confianza del Rey y por ello pudo intervenir ante él en favor de su sobrina Bárbara Coronel, encarcelada en Guadalajara y condenada a muerte por haber tramado con un apuntador el asesinato de su marido (G, 422; AP1, 298) —sin embargo, sólo la <em><strong>Genealogía</strong></em>, a la que sigue Aguilar Priego, recoge este parentesco de Cosme Pérez con Bárbara Coronel—. Existen otros testimonios que hablan del aprecio del Rey y de la familia real por Cosme Pérez. Así, M. L. Lobato recoge el testimonio del entremés de Avellaneda <em><strong>La portería de las damas</strong></em>, probablemente escrito hacia 1650, en el que se deja constancia de este aprecio, haciendo referencia al hecho de que el autor Pedro de la Rosa hubiera quitado a 'Juan Rana' los papeles de representar porque el gracioso había perdido la memoria, y la tristeza que ello supuestamente habría ocasionado en la familia real: "Qué ha de hacer la infantica [María Teresa de Austria] / sin su Juan Rana? ¡Ay, bella chocotica! / y al rey, aunque lo encubre con el guante, / ¿quién le ha de hacer reír de aquí en adelante?" (Lo, 94; UT, I, 143).
<br>Según la <em><strong>Genealogía</strong></em>, fue "honbre exemplar de vida y quando estaba enojado con alguno no le llamaba por su nombre. Estábalo en una ocasión con Pedro de la Rosa y para nombrarle repetía muchas vezes: aquel que güele". La misma fuente refiere que en otra ocasión un poeta escribió un entremés en que aludía a la propiedad de Cosme Pérez de unas casas en la calle de Cantarranas, junto a una pastelería, y éste le pidió que lo omitiese, pues había solicitado al rey una ayuda que no le sería concedida si se supiera que tenía casas en propiedad (G, 117).
<br>Por otro lado, siguiendo con las anécdotas que se le atribuían a 'Juan Rana', Pellicer y después Cotarelo recogen el testimonio de Juan Caramuel, en su obra <em><strong>Primus calamus</strong></em> (Roma, 1663-1665, 2 tomos), de donde traducen la siguiente anécdota, referida a 'Juan Rana': "En Madrid, el Salón del Palacio, llamado el Bello (o Buen) Retiro, donde se representan las comedias, tiene alrededor algunas ventanas o balcones, que corresponden a los aposentos donde se sientan los Grandes cuando se representan comedias. En cierto entremés, en que Juan Rana el gracioso más festivo que conoció España, haciendo el papel de Alcayde de aquel Palacio, introduxo dos forasteros, a quienes mostró todo lo que había digno de verse en él; y quando llegó a mostrarles el teatro colgado, como se ha dicho, de preciosas pinturas entre las ventanas o balcones, les dixo: -Este es el Salón donde se canta y representan las Comedias. El Rey y la Reina se sientan allí; aquí los Infantes; los Grandes en aquella parte. Y volviéndose a mirar las ventanas donde había dos Señoras de la primera grandeza, les dixo: -Contemplad aquellas pinturas, qué bien y qué al vivo están pintadas aquellas dos viejas, no les falta más que la voz, y si hablasen creería yo que estaban vivas, porque con efecto el Arte de la pintura ha llegado a los sumo en nuestro tiempo. Esto dixo porque las Españolas se untan las caras, y se las pintan con varios ingredientes y menjurges; y así con razón aseguró Juan Rana que aquellas dos Señoras Grandes eran dos propísimas pinturas" (P, II, 69-71; CM5, CLXIII). 
<br>A su talento cómico, especialmente en el papel de alcalde rústico, se alude en el entremés de Luis de Belmonte <em><strong>La maestra de gracias</strong></em> ("si hace algún alcalde simple / que haya sobrado a Juan Rana, / a quien ciertos entremeses / perpetuaron la vara […]"), y en la segunda parte del baile <em><strong>El poeta de bailes</strong></em>, de Luis Quiñones de Benavente, "Cosme" hace referencia a su papel preferido, el de alcalde, y a sus atributos característicos: su "montera" y su "vara". Sin embargo, como observa H. Bergman, las obras breves testimonian también la actuación por parte de Cosme Pérez en otros papeles, por ejemplo el de médico, así en <em><strong>El doctor Juan Rana</strong></em>: "Tan ligero soy de cholla, / señores, que me he pasado / desde el tribunal de alcalde / al de médico, de un salto" (Be, 132-33).
<br>En algunos entremeses de Quiñones de Benavente —véase más abajo para el listado completo de aquellos en cuya representación participó Cosme Pérez— Bergman nota una posible alusión al escándalo del que fue protagonista Cosme Pérez en 1636 cuando fue acusado de pecado nefando (Be, 521-22): véase el año correspondiente. Posteriormente otros investigadores han insistido en este aspecto de la personalidad de Juan Rana, especialmente F. Serralta, que analizó un corpus amplio de piezas escritas para este actor señalando posibles alusiones a su supuesta condición homosexual y destacando la posibilidad de que determinados rasgos de la "máscara" cómica creada por el actor pudieran fundarse en una mímica específica en relación con esta condición. Dicho investigador lanzó al mismo tiempo la hipótesis de que su apodo de 'Rana' estuviese relacionado con cierta ambigüedad sexual que lo caracterizaría (véanse Frédéric Serralta, <em><strong>Juan Rana homosexual</strong></em>, en <em><strong>Criticón</strong></em>, 50 (1990), pp. 81-92, y <em><strong>La risa y el actor: el caso de Juan Rana</strong></em>, en I. Arellano, V. García Ruiz y M. Vitse (eds.), <em><strong>Del horror a la risa: los géneros teatrales clásicos. Homenaje a Christiane Faliu-Lacourt</strong></em>, Kassel, Reichenberger, 1994, pp. 287-302, así como Peter Thompson, <em><strong>Juan Rana, a Gay Golden Age Gracioso</strong></em>, en E. H. Friedman, H. J. Manzari y D. D. Miller (eds.), <em><strong>A society on Stage: Essays on Spanish Golden Drama</strong></em>, New Orleans, University Press of the South, 1998, pp. 239-52). Sin embargo, F. Sáez Raposo desliga la génesis tanto del nombre como del apellido del personaje de un factor concreto y singular, y la establece a partir de componentes diversos: de carácter folclórico, relacionados con las características físicas del actor e incluso relativos al posible timbre de voz que éste pudo tener. Según dicho investigador, esa presunta homosexualidad sería un elemento del que se supo obtener un alto rendimiento cómico al relacionarla con características ya fijadas en el personaje, pero que difícilmente se podría establecer como elemento fundador de su denominación (F. Sáez Raposo, <em><strong>El origen del nombre artístico Juan Rana</strong></em>, en <em><strong>Voz y Letra</strong></em>, XV, II, 2004, pp. 3-58).
<br>Según relata la <em><strong>Genealogía</strong></em>, "estubo retirado mucho tiempo por su edad y después de algunos años mandaron los Reies que saliera a una fiesta del Retiro el año [dejado en blanco] y le sacaron en un carro" (G, 117). Varey y Shergold, en su edición de la <em><strong>Genealogía</strong></em>, relacionan este episodio con el estreno de la obra de Calderón <em><strong>Fieras afemina amor</strong></em>, en la que Cosme salió en un carro para actuar en el entremés de <em><strong>El triunfo de Juan Rana</strong></em> (G, 117 n. 1). Varey y Shergold recogen el juicio de Cotarelo y Mori, quien supuso primero que la obra fue estrenada en 1668 (CM5, CLX), aunque en otro lugar afirmó que la obra fue estrenada con motivo de la celebración del cumpleaños de la reina madre, Mariana de Austria, que era el 22 de diciembre de 1669 pero que se aplazó hasta enero de 1670 (CM9, X 8-9). Sin embargo, documentos descubiertos con posterioridad y publicados por Varey y Rich Greer indican que el estreno de <em><strong>Fieras afemina amor</strong></em> de Calderón de la Barca se produjo en realidad el 29 de enero de 1672, y por tanto es ésta también la fecha de la representación del entremés (F, XXIX, 42). Como evoca, por su parte, M. L. Lobato, el lujo de esta fiesta provocó violentas reacciones, entre las que se incluyeron pasquines como el titulado <em><strong>A la fiesta que hizo en el Retiro a los Reyes el Príncipe de Astillano en 29 de enero de 1672</strong></em>, donde se aconsejaba al Rey elegir mejor a su privado y donde se citaba también a 'Juan Rana': "La milicia castellana, / para bencer en la lid, / solía sacar al Zid, / y ahora sale Juan Rana, / que después de sepultado / a las cosas de onor, / como al buen Cid Campeador / le tienen empapelado" (Lo, 109).
<br>Según F. Sáez Raposo, la primera aparición escénica del personaje "Juan Rana" habría tenido lugar hacia 1624, cuando participó en la representación de la comedia de Lope de Vega <em><strong>Lo que ha de ser</strong></em>. Dicho investigador señala que la participación del actor en esta obra es mínima, pero que aparecen ya bien definidos los dos pilares en los que estará asentada su caracterización futura: su profesión de alcalde y su marcada estulticia. El mismo investigador señala el lustro comprendido entre 1625 y 1630 como fecha en la que participaría también en la representación de la comedia de Juan Pérez de Montalbán titulada <em><strong>El segundo Séneca de España, don Felipe II</strong></em>, donde Cosme Pérez tuvo un papel de mucha mayor relevancia que el anterior, no sólo en cuanto a su extensión, sino en su implicación en la trama. A partir de las palabras pronunciadas por el personaje de "Bartola, sobrina de Rana" en el momento en que le anuncia y presenta en escena, Sáez Raposo deduce que el actor ya era bien conocido por el público, lo que no podría deberse sólo al pequeño papel interpretado en la citada comedia de Lope: "Dentro Juan Rana. / Rana.- Verá el diabro, o mala rabia / te dé, pues miren a donde / vino a caer. Si una tranca / enquillotro, arre. Rey.- ¿Qué es esso? / Bartola.- Mi tío. Rey.- Cómo se llama. / Bartola.- Juan Rana, y es Alcalde, / que antaño salió por Pascua" (F. Sáez Raposo, <em><strong>Juan Rana y el teatro cómico breve del siglo XVII</strong></em>, op. cit., 25-26).
<br>Como observó Bergman, con el paso del tiempo la máscara de "Juan Rana" se incorporó al repertorio general de tipos cómicos, pudiendo ser interpretada por otros actores, como se hace evidente en un pasaje de <em><strong>El caballero novel</strong></em>, pieza de autor desconocido: "...que no me persigáis en / los alcaldes, los Juan Ranas, / los doctores, los letrados, / los veinticuatro de chanza" (Be, 136).
<br>El nombre de Cosme Pérez aparece en los repartos de algunos manuscritos mencionados en las fuentes que manejamos:
<br>1) Así, en la Biblioteca Nacional de Madrid se conserva, con la signatura Res. 109, el manuscrito autógrafo de <em><strong>El desdén vengado</strong></em>, de Lope de Vega, fechado en Madrid el 4 de agosto de 1617. El manuscrito lleva licencias de representación fechadas en Madrid el 9 de septiembre de 1617 y en Lisboa el 21 de marzo de 1622. El reparto que figura en el manuscrito, que hemos consultado, podía corresponder, como ya indicó Rennert, a la compañía de Juan Bautista Valenciano (q. v.) y en él figura "Cosme" —probablemente Cosme Pérez— en el papel de "Leonardo" (Res. 109; R, 369, 553; CM5, CLVIII). 
<br>2) Igualmente, en la Biblioteca Nacional de Madrid se conserva, con la signatura Res. 84, el manuscrito autógrafo de <em><strong>La nueva victoria de don Gonzalo de Córdoba</strong></em>, de Lope de Vega, fechado en Madrid el 8 de octubre de 1622. El manuscrito lleva licencia de representación fechada en Madrid el 21 de octubre de 1622. El reparto que figura en el manuscrito, que hemos consultado, correspondería, como ya indicó Rennert, a la compañía de Juan Bautista Valenciano y en él figura "Cosme" —como ya supuso Rennert, se trataría de Cosme Pérez— en el papel de "El Capitán Medrano" (Res. 84; CM5, CLVIII; R, 371, 553, 614; Gr6, 356) —Cotarelo observó que los papeles representados por Cosme Pérez en las dos obras de Lope a las que acabamos de aludir no eran cómicos, y de ello dedujo que en esas fechas tempranas de su profesionalización su especialización como gracioso no se encontraba todavía definida. Dicho investigador sugirió que quizá aprendió su papel de gracioso de Tomás Fernández de Cabredo, el autor con el que trabajó años más tarde y que era especialista en el mismo (CM5, CLVIII)—.
<br>3) Además, el nombre de "Cosme", probablemente Cosme Pérez, aparece tachado en el reparto incluido en el manuscrito de <em><strong>La infelice Dorotea</strong></em>, de Andrés de Claramonte, conservado en la Biblioteca Nacional de Madrid. La comedia fue escrita en 1620 para la compañía de Juan Bautista Valenciano (q. v.) y el manuscrito contiene una licencia de representación fechada en diciembre de 1621, y otra fechada en Madrid el 12 de noviembre de 1622 (R, 370; Ch. Ganelin (ed.), <em><strong>La infelice Dorotea</strong></em>, Londres, Tamesis Books, 1987).
<br>4) Por otro lado, en una copia manuscrita que se conserva en la Biblioteca Nacional de Madrid (Ms. 15. 633) que contiene los dos primeros actos de <em><strong>Bien vengas mal</strong></em> [o <em><strong>Bien vengas mal si vienes solo</strong></em>] aparece bajo el nombre de Calderón, como autor de la obra, el de Antonio de Escamilla, a quien probablemente perteneció esta copia. El manuscrito incluye un reparto de actores en el que figura "Cosme", probablemente Cosme Pérez, en el papel de "Espinel". Cotarelo, al dar cuenta de este reparto, supuso que era anterior a su posesión por parte de Escamilla y, sin dar razones fundadas, identificó al autor de esta formación con Roque de Figueroa (CM9, IX, 178). Años más tarde, Varey y Shergold, al dar cuenta de este manuscrito, pusieron en duda también que Antonio Escamilla fuese el autor de la formación a la que corresponde este reparto, aunque sin mencionar la atribución a Roque de Figueroa hecha por Cotarelo (SV2, 277). No hemos podido identificar al autor de esta compañía, pero véase para el reparto completo la entrada de Antonio de Escamilla.
<br>Por su parte, Luis Quiñones de Benavente escribió varias piezas para Cosme Pérez, en cuyos repartos aparece bien bajo el nombre de Cosme Pérez o de su apodo 'Juan Rana' (Be, 522; G, 117). La lista de obras de Quiñones o atribuidas a este dramaturgo en las que figura el actor que nos ocupa son:
<br>1) La primera y la segunda parte del entremés <em><strong>El guardainfante</strong></em>, donde figura como "Juan Rana", representado por Tomás Fernández, según Bergman probablemente en otoño de 1634 (Be, 307, 522).
<br>2) El entremés cantado <em><strong>El soldado</strong></em>, en cuya reparto figura como "Cosme [Pérez]", representado por Tomás Fernández, según Bergman probablemente entre 1634 y 1635. Dicha investigadora señala que esta pieza figura también con los títulos de <em><strong>El mundo</strong></em> y <em><strong>El mundo al revés</strong></em> (Be, 356, 522, 423-24; SA, 312 n. 12).
<br>3) El reparto del entremés <em><strong>El doctor Juan Rana</strong></em>, en cuyo reparto figura como "Juan Rana", representado, según la hipótesis de Bergman, por Tomás Fernández (q. v.) y no por Pedro de la Rosa —como habían supuesto Sánchez Arjona y Rennert (R, 554; SA, 330)—, probablemente en 1635 (Be, 285, 522). En relación con este entremés, Cotarelo recuerda que uno de los temas de concurso de la academia burlesca celebrada en el Buen Retiro durante las fiestas de Carnaval de 1637 fue "doce redondillas digan la razón por qué las beatas no tienen unto, y si basta la opinión del Doctor Juan Rana para que se crea". Aunque parece aludirse al entremés <em><strong>El doctor Juan Rana</strong></em> de Luis Quiñones de Benavente, como en éste no aparece ninguna referencia al "unto de las damas", Cotarelo supuso que debe aludirse a otro entremés de Quiñones. En cualquier caso, según apunta dicho investigador, en este certamen disputaron el premio Luis Quiñones, que obtuvo el primero, y Juan Navarro de Espinosa, que obtuvo el segundo premio (CM5, CLVIII).
<br>4) El entremés <em><strong>El mago</strong></em>, en cuyo reparto figura como "Cosme [Pérez]", representado por Pedro de la Rosa —en cuya compañía trabajaba entonces probablemente Cosme Pérez— y Tomás Fernández, según Bergman probablemente en 1637 (Be, 316, 522). Por su parte, según M. L. Lobato, este entremés se representó en el Retiro durante la fiesta de San Juan de ese año (Lo, 101).
<br>5) El entremés <em><strong>Los muertos vivos</strong></em>, en cuyo reparto figura como "Cosme", representado por Pedro de la Rosa, según Bergman probablemente en 1636 (Be, 324, 522).
<br>6) El entremés <em><strong>El remediador</strong></em>, en cuyo reparto figura como "Cosme", representado por Pedro de la Rosa, según Bergman probablemente entre 1636 y1640 (Be, 344, 522). 
<br>7) El entremés atribuido a Quiñones <em><strong>Pipote en nombre de Juan Rana</strong></em>, en cuyo reparto figura como "Cosme Pérez", que fue escrito, según Bergman, probablemente entre 1634 y 1640 (Be, 432, 522).
<br>8) El baile atribuido a Quiñones <em><strong>El poeta de bailes y el letrado. Segunda parte</strong></em>, en cuyo reparto figura "Cosme [Pérez]", representado, según Bergman, probablemente entre 1634 y 1640 por las compañías de Tomás Fernández o Pedro de la Rosa (Be, 438, 522).
<br>9) El baile atribuido a Quiñones <em><strong>Al cabo de los bailes mil</strong></em>, pieza que, según Bergman, parece ser una combinación de dos obras independientes basada en el personaje principal, que debió de ser "Juan Rana" (Be, 366-67, 522-23).
<br>Por otro lado, Antonio de Solís compuso también algunas de sus obras para Juan Rana, entre ellas el entremés <em><strong>Los volatines</strong></em>, que el actor representó, según M. L. Lobato, hacia 1655 con motivo de la fiesta que la Infanta María Teresa de Austria mandó hacer en el Salón del Palacio de Buen Retiro. Junto con Juan Rana representaron el entremés, según la misma investigadora, los siguientes actores: Bernarda Manuela, 'la Coronela' [Ana Coronel], 'la Borja' [Mariana de Borja] y Mateo Godoy. En la obra se testimonia el atractivo que para el público suponía la presencia del actor en una compañía, hasta el punto de que en este entremés el personaje de Bernarda bromea con la utilización del nombre del actor como cebo para el público en los carteles que anunciaban las comedias en los teatros: así, uno, "para juntar mucho pueblo / ponía que con Juan Rana / servía un autor, y luego / acabada esta comedia / el otro ponía lo mesmo" (Lo, 94-95).
<br>Según apunta M. L. Lobato, Antonio de Solís contó también con Juan Rana para el entremés titulado <em><strong>La Zarzuela</strong></em>, el cual, según el manuscrito que se conserva de la obra, se hizo en el Palacio de la Zarzuela con motivo de una "Fiesta que se representó a Su Majestad" en un día de mucho frío, según se indica en el primer verso del entremés (Lo, 97). Según recuerda M. L. Lobato, Solís también escribió para el actor que nos ocupa el entremés <em><strong>El retrato de Juan Rana</strong></em>, pieza totalmente diferente, según afirma dicha investigadora, del que años más tarde hizo Calderón con el mismo título. En él actuó también Bernarda Ramírez. La pieza se representó en palacio y M. L. Lobato deduce, a partir del hecho de que en las alabanzas finales se omite cualquier referencia al príncipe Felipe Próspero, que el entremés fue anterior a su nacimiento, que tuvo lugar en noviembre de 1657 (Lo, 97). Además, Juan Rana intervino en la representación de otras piezas de teatro breve de Solís, como el sainete <em><strong>Las fiestas bacanales</strong></em>: véase más abajo el año 1655. Asimismo, intervino en la representación del entremés <em><strong>Salta en banco</strong></em>, del entremés <em><strong>El niño caballero</strong></em> y del sainete <em><strong>Aguardad, supremos dioses</strong></em>, que se representaron junto a la comedia <em><strong>Triunfos de Amor y Fortuna</strong></em>, todos ellos obra del mismo Solís: véase más abajo el año 1658.
<br>Según indica M. L. Lobato, 'Juan Rana' también actuó en el entremés <em><strong>El ayo</strong></em> de Agustín Moreto, representado con motivo del cumpleaños del rey Felipe IV, quizá, como propone la citada investigadora, en el año 1662, en el que actuaron también Mateo de Godoy y Jerónimo Morales (Lo, 107). En el entremés <em><strong>La loa de Juan Rana</strong></em>, escrito también por Moreto, intervino como actor Cosme Pérez, y en un parlamento entre [Miguel de] Orozco y [Juan] Rana se alude a la reciente viudez de [Sebastián de] Prado y a su encarcelamiento, que impedía su participación en los festejos de Palacio, motivo por el cual Rana, ante la falta de éste y otros actores, proponía imitar a diversos actores de su tiempo como [Antonio de] Escamilla, [Alonso de] Olmedo, [Mateo de] Godoy, María de Quiñones, María de Prado y Manuela Escamilla (Lo, 106; CM1, 152). Según Cotarelo, esta loa se representó ante los Reyes el 22 de diciembre de 1662, dado que la muerte de Bernarda Ramírez, mujer de Prado, se había producido el 24 de octubre de 1662 y se sabe por un testimonio del 17 de diciembre de 1662 que Prado estaba encarcelado y por ello no podría participar en la fiesta palaciega que había de tener lugar cinco días después (CM1, 152). M. L. Lobato, sin embargo, sugiere que el entremés se representó "un 26 de julio", festividad de Santa Ana, pues en los últimos versos se hace mención del objeto de la fiesta, festejar el nombre de Ana (Lo, 106) —sin embargo, parece más probable la fecha propuesta por Cotarelo, pues a sus argumentos hay que añadir que el 22 de diciembre era la fecha en que se celebraba el natalicio de la reina Mariana de Austria—.
<br>Aparte de las piezas breves mencionadas a lo largo de la presente entrada, Cotarelo ofreció un primer corpus de los entremeses que diferentes poetas escribieron para este actor o su personaje, corpus que ha sido actualizado y completado modernamente por Sáez Raposo, quien repasa también algunos problemas de atribución de estas obras. Dicho investigador recoge un total de setenta y tres piezas en un repertorio que se extiende a lo largo de más de tres siglos y medio —para el listado completo de estos entremeses, véase dicha fuente— (CM5, CLXIII; F. Sáez Raposo, <em><strong>Juan Rana y el teatro cómico breve del siglo XVII</strong></em>, op. cit, pp. 65-71).<br><br><div id="anyo-1617"><b>1617</b><br> </div>Según Cotarelo (CM5, CLVIII), a quien sigue Sáez Raposo (<em><strong>Juan Rana y el teatro cómico breve</strong></em>, op. cit., p. 23), Cosme Pérez formaba parte este año de la compañía de Juan Bautista Valenciano —pero esta noticia procede del reparto de la compañía de Juan Bautista Valenciano incluido en el autógrafo de <em><strong>El desdén vengado</strong></em>, de Lope de Vega, del que hemos dado cuenta en la introducción de la entrada, y es posible que dicho reparto corresponda al año 1622—.<br><br><div id="anyo-1621"><b>1621</b><br> </div>Consta la lista de la compañía de Juan Bautista Valenciano (q. v.) para representar el Corpus de este año en Sevilla, entre cuyos miembros figura Cosme Pérez como bailarín y representante (Se, 1242).<br><br><div id="anyo-1624"><b>1624</b><br> </div>Cosme Pérez aparece en la nómina de la compañía de Antonio de Prado (q. v.) incluida en un concierto, fechado en Madrid el 18 de marzo, entre dicho autor y la Villa de Madrid sobre la representación de dos autos para la fiesta del Corpus, en el que se establecía, entre las condiciones, la entrega que haría la Villa a cada actor de una vela de cera de media libra (PP, I, 205-06; CM5, CLVIII; CM1, 434).<br><br><div id="anyo-1634"><b>1634</b><br> </div>Según Cotarelo, Cosme Pérez pertenecía a la compañía de Tomás Fernández Cabredo, quien le envió este año a Madrid con la limosna de su compañía para la Cofradía de Nuestra Señora de la Novena, según se hacía constar en los libros de la Cofradía (CM5, CLVIII); el primer cabildo de la Cofradía de Nuestra Señora de la Novena al que asistió Cosme Pérez, según la <em><strong>Genealogía</strong></em>, fue el celebrado el 25 de junio (G, 117); en el archivo de la Parroquia de San Sebastián de Madrid, según Cotarelo, constaba la defunción, el 18 de septiembre, en la Calle del Niño, donde vivía, de un hijo de Cosme Pérez —probablemente de muy corta edad, según Cotarelo. Es posible que éste fuese también hijo de María de Acosta— (CM5, CLVIII).<br><br><div id="anyo-1635"><b>1635</b><br> </div>Según Cotarelo, este año Cosme Pérez volvió a entregar una limosna a la Cofradía de Nuestra Señora de la Novena, aunque no explicita si en este caso también lo hizo en nombre de la compañía de Tomás Fernández (CM5, CLVIII); el Libro de la Hacienda de la Cofradía de Nuestra Señora de la Novena lo menciona este año como mayordomo de la Cofradía (G, 117); según un documento fechado en Valladolid el 9 de marzo, Pedro, hijo de los actores Pedro de la Rosa y Catalina Nicolás, su mujer, que vivían en la calle de las Comedias, fue bautizado en la iglesia de Nuestra Señora de San Lorenzo de Valladolid, actuando como padrinos los actores Antonio de Vitoria y Josefa Román, y como testigos los actores Cosme Pérez y Antonio Ramos (AC, 90).<br><br><div id="anyo-1636"><b>1636</b><br> </div>En el Libro de la Hacienda de la Cofradía de Nuestra Señora de la Novena se hacía constar el gasto por el entierro de su mujer, María de Acosta (G, 117) —aunque, según infiere Cotarelo de las anotaciones realizadas en el citado libro, debió de morir hacia 1632 (CM5, CLVIIII)—; consta un auto de los comisarios del Corpus, fechado en Madrid el 30 de enero, en el que se mencionaba a Cosme Pérez entre los miembros de la compañía de Tomás Fernández (q. v.) (SV1, 99; PP, II, 91); consta un concierto, fechado en Madrid el 16 de febrero, del actor Cosme Pérez para representar "la parte principal de la graciosidad" en la compañía de Pedro de la Rosa, por lo que recibiría 10 rs. de ración, 20 rs. por cada representación, 50 dcs. por la fiesta del Corpus y tres caballerías para los viajes (PP, I, 244-45); según consta en una carta que refiere sucesos del 22 al 29 de noviembre, en esos días 'Juan Rana', que había sido detenido acusado de "pecado nefando", fue liberado. El escándalo empezó hacia el 18 de octubre con la detención de un tal Agustín Merlo por mantener una reyerta con el hijo de un músico de la capilla real llamado Valdés, quien lo acusó de haberle requerido sexualmente. Sometido a tormento, aquél acusó a otros —en muchos entremeses de Luis Quiñones de Benavente se hace referencia a este escándalo: véase lo dicho en la introducción de la entrada— (Be, 522, 325; CM5, CLVIII).<br><br><div id="anyo-1637"><b>1637</b><br> </div>Consta un concierto, fechado en Valencia el 3 de marzo, entre Juan de Cabanillas, procurador de Pedro de la Rosa, y el clavario del Hospital General de Valencia para que Rosa, estante en Madrid, acudiera a Valencia con toda su compañía, teniendo que traer sobre todo a Francisco de Velasco y a Cosme Pérez, apodado 'Rana', hacia el día 10 de julio, para efectuar cincuenta representaciones seguidas —para más detalles del concierto, véase la entrada del autor— (E, 509); según se recoge en los libros del clavario del Hospital General de Valencia, el 5 de diciembre no pudo representar en Valencia la compañía a la que pertenecía Juan Rana [la de Pedro de la Rosa] "per estar purgat Juan Rana" (JM, 542).<br><br><div id="anyo-1638"><b>1638</b><br> </div>Consta una escritura, fechada en Valencia el 10 de marzo, por la que el clavario del Hospital General de Valencia concertó con el autor Pedro de la Rosa que éste acudiría a Valencia con toda su compañía, que se hallaba en Zaragoza, y sobre todo con Francisco Velasco, Cosme Pérez, apodado 'Rana', la Góngora [Isabel de Góngora], que representaba damas, Antonia [de Santiago], que representaba damas y bailaba, y la mujer del dicho Rosa [Catalina Nicolás] para hacer treinta representaciones seguidas, doce de ellas nuevas, empezando alrededor del 1 de junio —para más detalles del contrato, véase la entrada del autor— (E, 510).<br><br><div id="anyo-1639"><b>1639</b><br> </div>Juan Rana formó parte de la compañía de Pedro de la Rosa (q. v.) que representó en el Corpus de Sevilla (SA, 327-28, 330; Se, 1240) —Cotarelo, al tomar esta información de Sánchez Arjona, afirma que estuvo en Sevilla y Málaga (CM5, CLVIII), pero nada dice éste sobre una estancia en Málaga—.<br><br><div id="anyo-1641"><b>1641</b><br> </div>Cosme Pérez entregó una limosna a la Cofradía de Nuestra Señora de la Novena procedente de la compañía de Pedro de la Rosa, de la que formaba parte (CM5, CLVIII); la Cofradía de Nuestra Señora de la Novena encargó a Cosme Pérez "ordenar lo relativo a unos enterramientos que para actores había dejado el antiguo actor Rodrigo de Saavedra" (CM5, CLVIII).<br><br><div id="anyo-1642"><b>1642</b><br> </div>Según la <em><strong>Genealogía</strong></em>, Cosme Pérez estuvo este año representando en Valencia (G, 117; SA, 330). Por su parte, según supone Cotarelo —probablemente a partir de esta noticia de la <em><strong>Genealogía</strong></em>—, pasó la primera parte del año en Valencia, desde donde llevó a Madrid una limosna para la Cofradía de Nuestra Señora de la Novena procedente de la compañía de Pedro de la Rosa, de la que formaba parte (CM5, CLVIII) —sin embargo, se documenta la estancia de la compañía de Pedro de la Rosa, a la que al parecer pertenecía Cosme Pérez, en Valencia a partir de septiembre: véase la entrada correspondiente—; consta una orden de los comisarios de las fiestas del Corpus, fechada en Madrid el 3 de marzo, para que Antonio de Prado y Pedro de la Rosa (q. v.), autores de comedias, no salieran de Madrid hasta que hubieran dado las muestras de sus compañías y se hubiera señalado las que debían representar en las fiestas del Corpus. Esto se notificó el 4 de marzo a Cosme Pérez, miembro de la compañía de Pedro de la Rosa, y a los demás autores y actores (SV5, 34); consta una orden de los comisarios de las fiestas del Corpus, fechada en Madrid el 28 de marzo, para que Antonio de Prado y Antonio [sic, por "Pedro"] de la Rosa (q. v.), autores de comedias, dieran relación de los miembros de sus compañías, entre los cuales aparece 'Juan Rana', viudo, como miembro de la compañía de Pedro de la Rosa (SV5, 34-35; CM5, CLVIII).<br><br><div id="anyo-1643"><b>1643</b><br> </div>El 27 de marzo —según Davis y Varey, y no el 28 de febrero, como lee Pérez Pastor— Catalina de Nicolás y la Rosa, en virtud del poder del 17 de febrero otorgado por su marido, Pedro de la Rosa, contrató en Madrid a Cosme Pérez, desde la fecha del contrato hasta el día del Corpus, 4 de junio, y, si fuese necesario para el autor, para los días de la Octava del Corpus, representando "su parte de toda la graciosidad" en comedias, autos, bailes y entremeses, como lo había hecho "en otros años antezedentes a este asiento", por lo que cobraría 12 rs. de ración, 20 rs. por representación, 50 dcs. por el Corpus, las caballerías "que siempre se le an dado" para el transporte y 1.000 rs. de préstamo. En nombre de su marido, Catalina se comprometía asimismo a pagar 456 rs. del viaje de Murcia, de hato y personas, y las raciones de Cuaresma (PP, II, 123; F, XXXV, 185); según Cotarelo, la compañía de Pedro de la Rosa representó en Madrid los autos del Corpus [4 de junio] y en la lista de actores presentada a las autoridades figuraba Cosme Pérez, aunque al principio parece que no quería representar, según se infiere de la anotación hecha por Pedro de la Rosa al entregar la lista de su compañía: "Está indeciso, porque no quiere representar" (CM5, CLIX) —pero se trata de un error de Cotarelo, pues, como se verá a continuación, esta noticia correspondería realmente a 1644—.<br><br><div id="anyo-1644"><b>1644</b><br> </div>El 4 de febrero Pedro de la Rosa (q. v.) y Pedro Ascanio dieron la primera lista de sus compañías, en la que figuraba Cosme Pérez, "el gracioso Juan Rana", miembro de la compañía de Pedro de la Rosa (SV5, 44-45) —pero véase también la noticia del 19 de marzo—; consta una obligación, fechada en Madrid el 22 de febrero, de Francisco de Guzmán y Morales, autor de comedias, de pagar a Pedro Ortiz de Urbina, tesorero de la Cofradía de Nuestra Señora de la Novena, 120 rs. que le había prestado. El autor no sabía firmar y en su lugar lo hizo Cosme Pérez (PP, II, 128); el 19 de marzo, en Madrid, Pedro de la Rosa (q. v.) dio una nueva lista de su compañía a los comisarios de las fiestas del Corpus y en la misma figura Cosme Pérez, aunque el autor apuntó que "está indeciso porque no quiere representar" (SV5, 45); consta un decreto de los comisarios de las fiestas del Corpus, fechado en Madrid el 21 de marzo, por el que se ordenaba volver a notificar a Cosme Pérez, gracioso, y a Antonio Marín, gracioso, que no salieran de la Corte sin licencia o mandato de los comisarios, bajo pena de 200 dcs. y destierro del Reino, y a Antonio Marín se le embargaría su ropa. Se notificó también a Alonso de Paz, autor de comedias, y a su mujer, Manuela Bernabela, que no llevaran con su compañía a Antonio Marín fuera de Madrid, bajo pena de 1.000 dcs. y riguroso castigo. Pero al día siguiente de hacerse las notificaciones, Pedro de la Rosa, autor de comedias, dijo que consentía que Antonio Marín, gracioso en la compañía de Alonso de Paz, se fuera con ésta, pues él tenía concertado con Cosme Pérez, gracioso, que representara en su compañía durante las fiestas del Corpus y el resto del año, por lo cual pedía que se eliminase el embargo a la compañía del autor Alonso de Paz para que pudiera ir donde quisiera y llevar consigo a Antonio Marín, lo cual se hizo ese mismo día (SV5, 45); en la documentación relativa a las fiestas del Corpus madrileño se indica, con fecha del 15 de julio, Cosme Pérez, miembro de la compañía de Pedro de la Rosa, autor de comedias, recibió 50 dcs. de ayuda de costa por haber representado allí durante las fiestas del Corpus y no haber ido a otra parte (SV5, 46); consta una obligación, fechada en Madrid el 14 de agosto, de Pedro de la Rosa y de los miembros de su compañía Cosme Pérez, Jerónimo de Morales, Juan de Tapia, Jacinto Becerril, Pedro de Contreras y Pedro de Cifuentes, de estar en Madrid el 1 de noviembre para representar en los corrales arrendados por Francisco Garro de Alegría, quien permitiría al autor ir durante este tiempo a representar a Salamanca o a otros sitios para poder cobrar lo que éste le debía (PP, II, 132; F, XXXV, 246).<br><br><div id="anyo-1646"><b>1646</b><br> </div>Consta una obligación, fechada en Zaragoza el 18 de febrero, por la que el autor Pedro de la Rosa prorrogaba el contrato del actor de su compañía Cosme Pérez, vecino de Madrid y residente en Zaragoza, hasta el día del Corpus de este año y las Octavas. Cosme Pérez continuaría recibiendo diariamente 32 rs., moneda jaquesa de Aragón, por representación. Una vez terminada la temporada teatral, Cosme Pérez podía decidir libremente marcharse de la compañía o permanecer en ella hasta cumplir un año de contrato sin que el autor pudiera "descontarle ni hacerle pagar cantidad alguna por razón de las raciones que le huviere dado en la Quaresma del presente año". El autor Pedro de la Rosa se comprometía a "volverle a la villa de Madrid" desde el lugar donde terminasen la temporada, así como a darle tres caballerías y pagarle los gastos del transporte de su ropa (GH1 41-42; GH, 227, 194, 121-22).<br><br><div id="anyo-1648"><b>1648</b><br> </div>Consta un poder, otorgado en Madrid el 22 de marzo, de Antonio García de Prado, autor de comedias, a favor de Cosme Pérez y Francisco Ortiz, miembros de su compañía, para que concertara fiestas en su nombre (F, XXXV, 296); en una carta, fechada el 9 de junio de este año, de Felipe IV a la condesa de Paredes de Nava, doña Luisa Enríquez Manrique de Lara, que había ingresado este año en el Convento de San José de Malagón, el Rey escribía: "Os hecho mucho de menos para los autos [del Corpus de este año], que me dicen que vuestro amigo Juan Rana hace famoso papel, y se me acuerda de lo que os hacía reír" (Lo, 82); en otra carta de Felipe IV a doña Luisa Enríquez Manrique, fechada el 7 de julio, el Rey, refiriéndose a Diego de Martos, que fue con recados de palacio al convento en donde estaba doña Luisa, escribe: "A su vuelta [de Diego de Martos] os daría nuebas de Juan Rana pues se rió harto con él, y justamente porque estubo del buen humor que vos sabéis, y Bernardilla [¿Bernarda Ramírez?] no lo descuida […] Aora hacen los autos en el corral y acude mucha gente a ellos con la buena gana que tienen de visitar aquella santa cassa" (Lo, 82).<br><br><div id="anyo-1649"><b>1649</b><br> </div>En una carta escrita desde San Lorenzo de El Escorial el 18 de octubre de este año a doña Luisa Enríquez Manrique de Lara, Felipe IV escribía: "tenemos por güesped a Juan Rana, que siempre está de tan umor como le dejastes" (Lo, 80); el 15 de noviembre, día en que Mariana de Austria hacía su entrada en Madrid como nueva reina, Cosme Pérez, por mediación de su personaje, le hizo llegar un memorial titulado <em><strong>Memorial que dio Juan Rana a la Reina Nuestra Señora Doña Mariana de Austria en su entrada en esta corte antes de llegar a Palacio</strong></em>, en el que solicitaba la concesión de algún oficio palaciego a su hija para, de esta manera, paliar su difícil situación económica y aumentar sus posibilidades de contraer matrimonio (F. Sáez Raposo, <em><strong>Juan Rana y el teatro cómico breve del siglo XVII</strong></em>, op. cit., p. 39).<br><br><div id="anyo-1650"><b>1650</b><br> </div>En una carta escrita por Felipe IV a doña Luisa Enríquez Manrique, el Rey se refería a las recientes fiestas de Carnaval en palacio, relatando la imitación que don Andrés Ferrer había realizado de 'Juan Rana': "Muy regocijadas Carnestolendas hemos pasado y la gente moza se ha divertido y entretenido. Harta soledad me hizo que no viésedes querer imitar a Juan Rana a don Andrés Ferrer, que cierto de puro frío nos hacía reír, pero la comedia fue buena, particularmente lo cantado y la representación de las mujeres, que Amarilis [María de Córdoba] salió a la luz y está tan gran farsanta como siempre". En esta misma carta, el Rey se refiere también a la presencia del embajador turco en Madrid, presenciando la comedia en palacio, y a su afición a la comedia y al trato con los actores, manifestada por el hecho de que, según contaba el Rey, se hubiese desplazado al Escorial llevándose consigo a 'Juan Rana' y a otro farsante que llamaban Mejía —Antonio Mejía, según la identificación de M. L. Lobato, quien supone que 'Juan Rana' debió de intervenir en la comedia mencionada y quizá estuvo alojado en palacio cuando el Rey le pidió a él y a Mejía que acompañaran al embajador al Escorial— (Lo, 87-88); en la memoria que presentó Antonio García de Prado (q. v.) para hacer los autos en las fiestas del Corpus en Madrid se incluye la lista de su compañía, en la que figura Cosme Pérez (SV5, 94; PP, 170; LB, 347; R, 554); Cosme Pérez 'Juan Rana' figura también entre los miembros de una lista de la compañía de [Antonio de] "Prado farsante" (q. v.) que apareció publicada por Cruzada Villaamil junto con varias noticias de representaciones particulares hechas al Rey en 1623 por la compañía de Antonio de Prado (CV, 8) —pero es imposible que esta lista sea de 1623, ya que incluye a José Antonio de Prado, que nació en 1635. Creemos que corresponde a la compañía de Antonio de Prado hacia 1650, pues muchos de los miembros que figuran en la lista publicada por Cruzada Villaamil sabemos que estaban en la lista de la compañía que Prado presentó para representar en el Corpus de Madrid en 1650, como es el caso de Cosme Pérez—; consta una obligación, fechada en Madrid el 8 de noviembre, de Antonio García de Prado (q. v.) y de los miembros de su compañía, entre ellos Cosme Pérez, para hacer cuarenta representaciones en Toledo, continuando después en la Corte hasta el miércoles de Ceniza del siguiente año, comprometiéndose también a preparar en Toledo dos comedias, <em><strong>Honrar bien, que Dios es Dios</strong></em> y <em><strong>El niño de Santa Bárbara</strong></em>, y en Madrid otra comedia nueva (PP, II, 141) —según Cotarelo, que manejó documentos de la Cofradía de Nuestra Señora de la Novena, los títulos de las dos comedias fueron <em><strong>Obrar bien, que Dios es Dios</strong></em> y <em><strong>El niño de Santa Barbara</strong></em>, ambas puestas en escena en Toledo por la compañía de Antonio de Prado (CM1, 455)—.<br><br><div id="anyo-1651"><b>1651</b><br> </div>Consta una comunicación, fechada en el real Sitio del Buen Retiro el 26 de abril, del Duque de Nájera al grefier de la Reina Simón de Alcántara, por la que le informa: "Su Magestad, Dios le guarde, por su orden escrita en Aranjuez, en respuesta a carta mía de 24 de este, es servido de hacer merced a Juan Rana de una ración ordinaria, que ha de gozar por la casa de la Reina [Mariana de Austria], nuestra Señora, en consideración de lo que la hace reír, y en esta conformidad se le apuntará en los libros para que se le acuda con ella desde la hecha de esta" (PP, II, 144; Lo, 80); en una carta de Felipe IV, fechada en Madrid el 6 de junio de este año, a doña Luisa Enríquez Manrique, el Rey reflejaba la expectativa generada en palacio ante la próxima representación de los autos del Corpus, que iban a permitir "ver a 'Juan Rana' en los autos, que dicen son muy buenos y que tiene buen papel" (Lo, 89); con motivo del parto del primer hijo de la reina Mariana de Austria, la infanta Margarita, y también con motivo del cumpleaños de la Reina y de su recuperación del "accidente que le sobrevino estando el Rey Nuestro Señor en las Descalzas, y con su presencia volvió del desmayo", se representó en palacio la comedia de Calderón <em><strong>Darlo todo y no dar nada</strong></em>, con una loa de Antonio de Solís, en la que participaron Mariana y Luisa Romero, 'la Borja' [Mariana de Borja], María de Quiñones, María de Prado y Bernarda Ramírez (Lo, 89; CM1 589) —según Cotarelo, para esta función se había reunido a las compañías de [Diego] Osorio y [Sebastián de] Prado y, durante esta misma fiesta, se representó el entremés <em><strong>El infierno de Juan Rana</strong></em>, en el que la actriz Bernarda Ramírez, célebre por imitar la figura de Juan Rana, interpretó el papel del "Alma" de este actor, quien también intervino en la representación (CM1, 589)—.<br><br><div id="anyo-1652"><b>1652</b><br> </div>Cotarelo se refiere a la existencia en el Archivo Municipal de Madrid de una carta, fechada el 5 de junio, de Juan González, hijo de José González, consejero y protector de los teatros, a don Francisco Sardeneta y Mendoza, solicitando en nombre de su padre una ayuda de costa de 400 rs. para Cosme Pérez 'Juan Rana' "porque necesita della" (CM5, CLIX) —Cotarelo infiere de esta carta que el actor tomó parte en las fiestas del Corpus—.<br><br><div id="anyo-1653"><b>1653</b><br> </div>En una carta, fechada en Madrid el 9 de junio de este año, Felipe IV daba cuenta a doña Luisa Enríquez Manrique de una comedia representada en el Coliseo del Buen Retiro en la segunda quincena de mayo: "Muy buena ha sido la comedia, y vuestro amigo 'Juan Rana' a cumplido famosamente con sus obligaciones; bien creo que si le hubiérades visto, se atreviera la risa a inquietar la función. La gente moza se ha entretenido harto" (Lo, 92-93) —la comedia representada fue <em><strong>Andrómeda y Perseo</strong></em>, de Calderón, en la que Cosme Pérez representó el papel de "Bato" (F. Sáez Raposo, <em><strong>Juan Rana y el teatro cómico breve del siglo XVII</strong></em>, op. cit., p. 43)—; entre la documentación relativa a las fiestas del Corpus de Madrid consta una lista de los gastos de las fiestas del Corpus de este año, en la que se indica que Cosme Pérez recibió 1.000 rs. (SV5, 110; CM5, CLIX); consta la respuesta, transmitida el 22 de noviembre por el Conde de Altamira, a la solicitud de "Cosme Pérez, alias Juan Rana" realizada el 18 de dicho mes, referente al "paso de una ración ordinaria que goza por la despensa de la casa de la Reyna nuestra señora, para una hija suya [Francisca María Pérez], con calidad que sea desde luego y que la susodicha no la pueda gozar si hubiese de andar en la farsa" (PP, II,153; Lo, 80-81).<br><br><div id="anyo-1654"><b>1654</b><br> </div>En una carta de Felipe IV a doña Luisa Enríquez Manrique, fechada el 17 de febrero de este año, el Rey se refería a a la participación de Juan Rana en las fiestas de Carnaval de aquel año, donde había divertido con su actuación a la Reina y a la infanta María Teresa (Lo, 94); a este mismo año corresponde la puesta en escena de <em><strong>El laberinto del mundo</strong></em>, de Calderón, único auto sacramental en el que se tiene constancia de que participara Cosme Pérez (María Luisa Lobato, "Fiestas teatrales al infante Felipe Próspero (1657-1661) y edición del baile <em><strong>Los Juan Ranas</strong></em> (XI-1658)", en <em><strong>Scriptura</strong></em>, 17 (2002), p. 259); consta un concierto, fechado en Madrid el 29 de junio, por el que Luis López [Sustaete] (q. v.) se comprometía a participar con su compañía, llevando a "Cosme Pérez para que haga la graciosidad", en las fiestas de Nuestra Señora de Agosto de Navalagamella (Madrid), en las que se harían tres comedias, una el día 14 y dos el 15, con bailes y entremeses (F, XXXVI, 376); consta un concierto, fechado en Madrid el 3 de julio, por el que la compañía de Luis López Sustaete (q. v.), "llevando a Cosme Pérez para la graciosidad", se concertaba con la Cofradía de Nuestra Señora del Rosario de Vicálvaro (Madrid) para participar en las fiestas de Nuestra Señora de Septiembre, en las que se harían dos representaciones de un auto y una comedia con bailes y entremeses (F, XXXVI, 375).<br><br><div id="anyo-1655"><b>1655</b><br> </div>Una actriz llamada Bernarda, identificada por Rennert con Bernarda Manuela y por Lobato con Bernarda Ramírez, intervino junto con Cosme [Pérez] y Francisca de Castro en la representación del sainete <em><strong>Las fiestas bacanales</strong></em>, escrito por Antonio de Solís para la versión palaciega de su comedia <em><strong>Eurídice y Orfeo</strong></em>, que se representó el 6 de febrero de 1655 —la comedia se había representado en Pamplona hacia 1643 con motivo del nacimiento de un hijo del Conde de Oropesa y fue refundida y convertida en zarzuela por el mismo Solís para ser representada en Palacio, y fue para esta ocasión, según afirma Lobato, que se escribió el sainete que la acompañó— (R, 433-34; Lo, 101; UT, II, 612, 613); el 7 de febrero, domingo de Carnaval, Juan Rana intervino en la loa de Antonio de Solís para la comedia <em><strong>Las amazonas</strong></em>, representada ante Su Majestad, junto con los actores Mateo de Godoy, María de Quiñones, 'la Borja' [Mariana de Borja], Luisa y Mariana Romero y Bernarda [Ramírez] —según apunta M. L. Lobato, la comedia se repitió en Carabanchel de Arriba en el mes de junio, con el autor Diego de Osorio al frente del grupo de representantes: pero esta representación en Carabanchel de Arriba tuvo lugar en 1656, y aunque la compañía que se hizo cargo fue la de Diego Osorio (q. v.), no sabemos si participaron los mismos actores— (CM1, 592; Lo, 94-95). Durante esta misma fiesta se representó el entremés <em><strong>Las visiones</strong></em>, en el que intervinieron 'Juan Rana' en el papel de "Juan Rana" y Bernarda Ramírez representando el papel de "Quiteria, mujer de Juan Rana" (CM1, 592); el martes de Carnaval de este año [9 de febrero] —aunque Lobato por error indica como fecha del representación el año 1656— se representó ante Sus Majestades la comedia <em><strong>Un bobo hace ciento</strong></em>, de Solís, en cuya loa alegórica Juan Rana hizo de "Tiempo", compartiendo escenario con Bernarda Ramírez, Luisa Romero y Mariana Romero, 'la Patata' [¿Antonia del Pozo 'la Patata' o Luciana 'la Patata'?], [Mariana de Borja] 'la Borja' y 'Poca Ropa' (Lo, 92). Según Cotarelo, 'Juan Rana' y Bernarda Ramírez representaron los papeles principales de la loa, mientras que Luisa y Mariana Romero, Luciana Leal y Mariana de Borja representaron las Cuatro Edades: de Oro, de Plata, de Cobre y de Hierro (CM1, 593); consta un contrato, fechado en Madrid el 24 de abril, por el que Diego Osorio (q. v.) se concertó con los comisarios de la fiesta de Señora Santa Librada de Sigüenza (Guadalajara) para participar con su compañía, excepto Cosme Pérez, en la fiesta de Santa Librada [20 de julio] de dicha ciudad (F, XXXVI, 398); entre la documentación relativa a las fiestas del Corpus madrileño consta un documento, sin fecha, en que se recogen los gastos de las fiestas del Corpus de Madrid de este año [27 de mayo], en el que se indica que 'Juan Rana' recibió 1.400 rs. (SV5, 119); según un testimonio recogido contemporáneamente por Jerónimo de Barrionuevo el 26 de junio, se representó en el Retiro el día de San Juan —probablemente el 23 de junio, noche de San Juan— la obra <em><strong>La restauración de España</strong></em> —también titulada <em><strong>La renegada de Valladolid</strong></em>, según Lobato (Lo, 96)—, que Barrionuevo describe del siguiente modo: "comedia burlesca. La primera jornada de Monteser; la segunda de Solís; la tercera, de don Diego de Silva, alias abad de Salas, hijo de la princesa de Mélito; el gracejo y sainetes de Cáncer; entremeses y danzas de otros ingenios selectos de la Corte. Setenta mujeres fueron las que la representaron, y Juan Rana tan solamente hizo el hombre y papel del Rey. 'La Romerilla' [¿Mariana o Luisa Romero?] salió en una hamaca a decir la Loa, y en un entremés donde se remedaba lo que pasa en el Prado aquella noche [de San Juan], entró un cochecillo pequeño en el salón alto donde se hacía, con cuatro mujeres en él y dos mulas que le tiraban, siendo otra mujer el cochero que le guiaba, subiéndole con una tramoya por las escaleras, como si lo hiciera en una de las calles del Prado" (Ba, I, 153; CM1, 595 nota; CM5, CLIX) —según Paz y Melia, en la Biblioteca Nacional de Madrid se conserva el manuscrito de <em><strong>La renegada de Valladolid</strong></em>, comedia burlesca, escrita por tres ingenios (Francisco de Monteser, Antonio de Solís y Diego de Silva), en la que se indica "Representose a Su Majestad en el Salón del Buen Retiro el día de San Juan, año de 1655". Esta comedia va precedida de una loa de don Antonio de Solís e incluye también el baile <em><strong>El contraste de amor</strong></em>, de Juan Vélez, al fin de la primera jornada, el baile entremesado de <em><strong>Los hombres deslucidos que se pierden sin saberse cómo ni cómo no</strong></em>, de don Jerónimo de Cáncer, al fin de la segunda jornada, y el entremés de <em><strong>La noche de San Juan, y Juan Rana en el Prado con escribano y alguacil</strong></em>, de don Jerónimo de Cáncer, al final de la tercera jornada (PM, 473, nº 3136)—; Cosme Pérez figura como testigo en el testamento de José Antonio Quevedo (q. v.), fechado en Madrid el 2 de septiembre (F, XXXVI, 408).<br><br><div id="anyo-1656"><b>1656</b><br> </div>En una carta de Baccio del Bianco, fechada [¿en Madrid?] el 3 de marzo de este año, el escenógrafo italiano describe pormenorizadamente una fiesta que tuvo lugar en el Salón del Palacio del Buen Retiro, organizada por la infanta María Teresa para festejar el restablecimiento de la Reina Mariana de Austria, en la que se representó la comedia <em><strong>Pico y Canente</strong></em> de Luis de Ulloa y Pereira y Rodrigo de Ávila, y en cuya loa y sainetes, escritos por Antonio de Solís, actuó 'Juan Rana', al que Bianco denomina "il Grazzioso della Regina". Según se relata en la carta, 'Juan Rana' aparecía en lo alto sobre una nube de oro, traído por "Flora", supuestamente trasladado desde el Parnaso para que representase la comedia. El prólogo acabó en un baile con castañetas, en el centro del cual estaba Juan Rana "con le sue buffonerie solite". Baccio describe también la loa, afirmando que cantaban en ella Luisa y Mariana Romero. Dio fin a la comedia el sainete en el que intervinieron 'Juan Rana', Luis de Mendoza y Luisa Romero. Es probable, como observa Lobato a partir de una alusión en el sainete final de Solís, que la fiesta tuviera lugar en febrero, durante el período de Carnaval —Cotarelo dató erróneamente esta fiesta y representación en 1653 (CM1, 589)—. Por otro lado, en la misma carta el escenógrafo italiano da cuenta de que la fiesta se representó después en el Coliseo, donde se repitió durante cuatro días, con gran asistencia de público, especialmente para la función que tuvo lugar el lunes de Carnaval (Lo, 90-92).<br><br><div id="anyo-1657"><b>1657</b><br> </div>El 17 de enero de este año 'Juan Rana' intervino en la comedia <em><strong>El golfo de las sirenas</strong></em>, de Calderón, y en la mojiganga <em><strong>Juan Rana en la Zarzuela</strong></em>, intrínsecamente vinculada a esta obra, que se representó ante Sus Majestades en el Real Sitio de la Zarzuela por las compañías de Diego Osorio y Pedro de la Rosa (Lo, 102); consta un auto, sin fecha pero anterior al 25 de enero, en el que los comisarios de la fiesta del Corpus de Madrid ordenaban que los autores Pedro de la Rosa (q. v.), Francisco García 'el Pupilo' y Diego Osorio, así como los miembros de sus compañías, no salieran de la Corte y estuvieran listos para que se formaran las compañías que habían de representar los autos, y que se les embargaran los vestidos y demás bienes. A partir del 9 de febrero se notificó a los representantes, entre los que se encontraba Cosme Pérez, "que se llama Juan Rana" (SV5, 125; PP, II, 167-68); en uno de sus <em><strong>Avisos</strong></em>, fechado el 31 de enero, Jerónimo de Barrionuevo recoge la siguiente noticia: "Dicen envían una compañía de comediantes selectos de todas las demás al señor don Juan de Austria, y entre ellos la Bernardilla [¿Bernarda Ramírez?], los dos hermanos Pradillos [¿Sebastián García de Prado y José Antonio García de Prado?] y el mejor gracioso [¿Cosme Pérez?], para que desde allí pasen a Francia a que los vea la Reina Madre, que se lo ha escrito con grandes instancias, movida de lo mucho que se los ha alabado allá los que estuvieron aquí en el Retiro, y que el viaje y galas costarán 50.000 ducados" (Ba, II, 58); Alenda y Mira recoge una relación en la que se da cuenta de la fiesta pública celebrada en Madrid el 6 de diciembre, con motivo de la salida de Su Majestad en público para dar gracias a la Virgen de Atocha por el nacimiento del Príncipe, y en ella consta que, para esta ocasión, "hubo arrimado a las rejas de palacio un tablado, y en él representando la compañía de [Pedro de la] Rosa […] Estava junto a Santa María otro tablado y en él representando la compañía de [Francisco García] 'Pupilo' con 'Juan Rana' tan gracioso como suele […] A la esquina de la cárcel de la Villa huvo hecho otro tablado, y en él estavan cantando con sonoros instrumentos las hijas de [Antonio de] Escamilla [Manuela, Ana y María]" (AM, 331; CM1, 608, n. 1; CM5, CLIX).<br><br><div id="anyo-1658"><b>1658</b><br> </div>Para celebrar el nacimiento del príncipe Felipe Próspero, el dramaturgo Agustín Moreto escribió el entremés <em><strong>El alcalde de Alcorcón</strong></em>, representado el 6 de enero de este año con motivo de la salida de la Reina a misa en la Real Capilla, en el que intervinieron, según M. L. Lobato, los actores Juan Rana, Bernarda Ramírez, Simón Aguado, Luis de Mendoza, Mariana y Luisa Romero, Mariana de Borja y Luciana Leal (Lo, 97-98); en una carta fechada el 12 de febrero de este año, el Rey Felipe IV escribía a doña Luisa Enríquez Manrique, condesa de Paredes de Nava, informándole sobre las fiestas de Carnaval en el Palacio del Buen Retiro: "Y aquí se pasaron bien estas Carnestolendas en que no dejara de hacer vuestro amigo Juan Rana muy bien su oficio" (Lo, 103 n. 64); el 27 de febrero de este año Sus Majestades asistieron en el Coliseo del Retiro a la representación de la comedia de Antonio de Solís <em><strong>Triunfos de Amor y Fortuna</strong></em>, que tuvo lugar con motivo de los festejos por el nacimiento del príncipe Felipe Próspero. Junto a la comedia se representó el entremés del mismo Solís titulado <em><strong>Salta en banco</strong></em>, en el que intervino Cosme Pérez junto a Bernarda [Ramírez], Bernarda Manuela, Luisa y Mariana Romero, [Francisca Bezón] 'la Bezona', María de Prado y María de Quiñones, y también se representó el entremés titulado <em><strong>El niño caballero</strong></em>, del mismo Solís, en el que actuó Juan Rana junto a Bernarda Ramírez, Francisca Bezona, [Mariana de Borja] 'la Borja', [Simón] Aguado, [Mateo] Godoy, [Pedro de la] Rosa y [Tomás de] Nájera. También se representó como final de fiesta el sainete <em><strong>Aguardad, supremos dioses</strong></em> del mismo Solís, en el que intervino Juan Rana junto a Bernarda Ramírez, Simón Aguado, Bernarda Manuela, Mariana Borja, Pedro de la Rosa, [Mateo] Godoy, [Diego] Osorio y [¿Jerónimo de?] Heredia (R, 433, 566; Lo, 98-99, 100-101, 103; UT, II, 612, 614) —a partir del entremés de <em><strong>El niño caballero</strong></em>, de Solís, Rennert creyó que podía deducirse que, en ese tiempo, Bernarda Manuela era la esposa de Cosme Pérez (R, 566), pero se trata de una deducción errónea: véase la introducción de la entrada—; entre la documentación relativa a las fiestas del Corpus de Madrid consta una memoria, sin fecha, de las "cosas de demasías" en los carros, en la que se indica que se pagaron 60 rs. por un vestido de gigante con su maza y cabeza para 'Juan Rana' [Cosme Pérez] y 36 rs. por otra cabeza de gigante con su tocado para la dama de 'Juan Rana' (SV5, 132). Posteriormente, en otra partida de gastos, figura un pago de 50 rs. para el "vestido de Juan Rana" (SV5, 135); consta una carta de pago, fechada en Madrid el 12 de octubre, de Cosme Pérez, vecino de Madrid y con domicilio en la calle de Cantarranas, en casas propias, que acusaba recibo de 784 rs. que había enviado Francisco Gutiérrez, yerno de Luis López Sustaete, desde Salamanca el 2 de octubre, y que eran a cuenta de lo que le debía a su suegro y del "precio de un vestido". Al encontrarse Luis López ya indispuesto para recibirlos y no estar ya "sujeto para recibirlos", Cosme Pérez se hizo cargo de ellos para pagar los gastos del entierro de López Sustaete en caso de que falleciese. En dicha carta de pago se declara que, aunque Gutiérrez había enviado 800 rs., Cosme Pérez recibió 16 rs. menos, que se emplearon en pagar al ordinario el porte del dinero (PP, II, 183; F, XXXVI, 460). Asimismo, consta el inventario de los bienes de Luis López Sustaete (q. v.) hecho por Juan Navarro Oliver y Antonio de Rueda, y que abarca una serie de documentos fechados en Madrid entre el 15 y el 21 de octubre. Entre los bienes se encuentran los 784 rs. de vellón enviados desde Salamanca a Luis López por Francisco Gutiérrez, su yerno, los cuales recibió Cosme Pérez, otro testamentario del dicho Luis López (F, XXXVI, 480); a finales de noviembre de este año se representó en palacio el baile de <em><strong>Los Juan Ranas</strong></em> para celebrar el primer cumpleaños del príncipe Felipe Próspero, que había nacido el 28 de noviembre de 1657, en el que actuaron Cosme Pérez, Bernarda Ramírez, Bernarda Manuela, Francisca Bezón, Gregorio de la Rosa, Simón Aguado, Diego Osorio, 'la Borja' [Mariana de Borja], Micaela Fernández, Ana —actriz cuyo apellido no se ha podido identificar—, María de Prado, Jerónima de Olmedo, Juana Caro, Micaela de Borja y una mujer que responde a la abreviatura 'Quir.' (Lo, 105; María Luisa Lobato, "Fiestas teatrales al infante Felipe Próspero (1657-1661) y edición del baile <em><strong>Los Juan Ranas</strong></em> (XI-1658)", en <em><strong>Scriptura</strong></em>, 17 (2002), p. 242, en donde M. L. Lobato corrige la fecha que figura en el primero de los dos artículos que acabamos de citar con la sigla Lo, 105).<br><br><div id="anyo-1659"><b>1659</b><br> </div>Según Cotarelo, a quien sigue M. L. Lobato, en una partida relativa a las fiestas del Corpus de este año se hace referencia a los 50 rs. que costó el vestido de Juan Rana (CM5, CLXI, Lo, 104) —pero se trata de un error, pues esta noticia es de 1658—.<br><br><div id="anyo-1660"><b>1660</b><br> </div>Según Urzáiz, este año Cosme Pérez representó junto con Bernarda Ramírez, 'la Quiñones' [María Quiñones] y 'la Borja' [María Borja] <em><strong>El baile perdido</strong></em>, de Antonio de Solís (UT, II, 612).<br><br><div id="anyo-1661"><b>1661</b><br> </div>Según refiere Lobato, el entremés <em><strong>El toreador</strong></em>, de Calderón, se representó ante la familia real con motivo del nacimiento del príncipe Carlos, el futuro Carlos II, y durante la representación de la obra, en la que participó 'Juan Rana', éste bajó del tablado para dirigirse al Rey, a la Reina y al "Principito", haciéndoles cortesías, y regresando después a ocupar su lugar en el tablado (Lo, 102).<br><br><div id="anyo-1662"><b>1662</b><br> </div>Según M. L. Lobato, 'Juan Rana' actuó en el entremés <em><strong>El ayo</strong></em> de Moreto, representado con motivo del cumpleaños del rey Felipe IV, quizá este año, como propone la citada investigadora, y en el que actuaron también Mateo de Godoy y Jerónimo Morales (Lo, 107).<br><br><div id="anyo-1663"><b>1663</b><br> </div>Según Cotarelo, el 1 de septiembre Catalina Ramos, hija de Juan Ramos y sobrina de "Cosme Pérez, comediante", murió en la calle de Cantarranas, donde vivía con su tío (CM5, CLX).<br><br><div id="anyo-1665"><b>1665</b><br> </div>Rennert y, después de él, Cotarelo recogieron el testimonio contemporáneo que aparece en las <em><strong>Memorias de Lady Fanshawe</strong></em>, quien relataba que el 5 de enero de este año, "entre las varias diversiones que hemos tenido estas Pascuas, fue la presencia, de Juan Araña [sic], el famoso representante, que trabajó aquí durante dos horas, con admiración de todos los que le vieron, considerando que tenía cerca de ochenta años" (R, 554; CM5, CLX; Lo, 108); en un escrito del Duque de Montalto dirigido al grefier, fechado en Madrid el 3 de junio, se le solicitaba que cumpliera lo que Su Majestad había mandado mediante un real decreto fechado el 31 de mayo: que la ración ordinaria que se había concedido al representante Cosme Pérez y que en su día fue cedida a su hija, Francisca María Pérez, volviera a ser disfrutada por dicho actor, con motivo del fallecimiento de ésta, hasta su muerte "por ser viejo y hallarse pobre" (PP, II, 198-99; Lo, 81).<br><br><div id="anyo-1666"><b>1666</b><br> </div>Según Cotarelo, el martes Santo [21 de abril], Cosme Pérez acudió a Palacio en representación de la Cofradía de Nuestra Señora de la Novena para recoger, como era habitual en estas fechas, la limosna para la capilla que los actores tenían en la madrileña iglesia de San Sebastián, recibiendo de la Reina un donativo de 150 rs., que Mateo Godoy recogió en los libros de cuentas de la Cofradía. En éstos figuran también los 16 rs. entregados por la Cofradía "a los silleros que llevaron a Cosme a Palacio para la demanda" —de esta circunstancia se ha deducido, a partir de Cotarelo, que por aquel entonces Cosme Pérez ya tendría problemas de movilidad debido a su avanzada edad— (CM5, CLX; Su, 79); el 13 de junio Cosme Pérez firmó en Madrid el primero de los dos testamentos que de él se conservan. En él declaraba que era vecino de esta ciudad, que vivía en "casas propias en la calle de Cantarranas, Parroquia de San Sebastián", que compartía de mala gana con un tal Justo de Lazárraga, y que se hallaba "con algunos achaques, aunque no me obligan a estar en cama". Entre otras estipulaciones referidas a las habituales peticiones de misas por su alma y detalles sobre sus honras fúnebres, pedía ser enterrado en el Convento de la Santísima Trinidad de Madrid, en la misma sepultura en que estaba enterrada su hija Francisca María Pérez. Entre las limosnas que entregaba figura una, de 200 dcs., para la Cofradía de Nuestra Señora de la Novena "para ayuda a las obras de su capilla". Asimismo, entregaba 200 dcs. a su sobrina María de Diego, que vivía en Tudela de Duero, 50 dcs. a su sobrino Francisco de Diego, que vivía en la villa de Olivares, y 200 dcs. para ayuda de un retablo "que me han dicho se trata de hacer", para la imagen de Nuestra Señora del Amparo, en la iglesia parroquial de Tudela de Duero. En el testamento se hace referencia a 10 dcs. que Antonio de Escamilla le debía por un vestido que le había vendido. Dejaba a sus albaceas y testamentarios el poder de vender sus bienes para cumplir las mandas de su testamento (Gr7, 660-62) —A. de la Granja, al publicar el testamento, observó la posibilidad de que el origen de Cosme Pérez estuviera en Tudela del Duero (Gr7, 659), circunstancia que, como referimos en la introducción de esta entrada, han confirmado investigaciones posteriores—.<br><br><div id="anyo-1667"><b>1667</b><br> </div>Igual que en el año anterior, Cosme Pérez acudió a Palacio para recibir el donativo de 150 rs. entregado por la Reina para la Cofradía de Nuestra Señora de la Novena, y esta vez los silleros que lo llevaron a Palacio recibieron 13 rs. (CM5, CLX).<br><br><div id="anyo-1668"><b>1668</b><br> </div>Como en los dos años anteriores, Cosme Pérez acudió a Palacio para recibir el donativo de 150 rs. entregado por la Reina para la Cofradía de Nuestra Señora de la Novena, y esta vez los silleros que lo llevaron a Palacio recibieron 14 rs. (CM5, CLX).<br><br><div id="anyo-1670"><b>1670</b><br> </div>Asistió al cabildo de la Cofradía de Nuestra Señora de la Novena celebrado el 2 de marzo (G, 117); el 9 de julio, en Madrid, firmó un nuevo testamento que invalidaba el anterior de 1666, en el que declaraba de nuevo ser vecino de Madrid, "que vivo en mis casas propias, en la calle de Cantarranas, Parroquia de San Sebastián, estando en pie levantado de la cama, aunque con algunos achaques", y pedía ser sepultado "sin ataúd en el convento de religiosos descalzos de la Santísima Trinidad desta Corte, en la sepultura que allí tengo mía propia, y donde está sepultada mi hija Francisca María Pérez". Aparte del dinero estipulado para misas por su alma y detalles de sus honras, entre las limosnas entregaba una de 150 dcs. de vellón para la Cofradía de Nuestra Señora de la Novena para ayuda de la fábrica de su capilla, y un pabellón de brocado que tenía en el oratorio de su casa. A su sobrina María de Diego y a su sobrino Francisco de Diego, ya mencionados en el testamento anterior, les enviaba 150 dcs. y 30 dcs. de vellón, respectivamente, y 150 dcs. de vellón para ayuda del retablo "que me han dicho se trata de hacer" para la imagen de Nuestra Señora del Amparo de la iglesia parroquial de Tudela del Duero. Asimismo, ordenaba pagar a su ama Isabel 150 rs. de vellón y lo que se le debiera de salario, y a Felipe Negro, su "paisano y pariente", otros 150 rs. y un espadín. Otorgaba a sus albaceas y testamentarios poder para rematar sus bienes y pagar todo lo estipulado en el testamento y, con el sobrante, pagar nuevas misas y limosnas para su alma y la de su hija y familiares (Gr7, 654-599) —A. de la Granja, al publicar el testamento, estimó que el desembolso en las misas previstas por su alma debió de rondar los 7.500 rs. y destaca la preocupación religiosa que manifiesta Cosme Pérez—.<br><br><div id="anyo-1672"><b>1672</b><br> </div>Constan los gastos relativos a "las galas" dadas a los representantes para vestir los papeles de la comedia <em><strong>Hércules</strong></em> o <em><strong>Fieras afemina amor</strong></em>, de Pedro Calderón de la Barca, que se tenía que representar en el Salón Dorado del Buen Retiro a costa del Príncipe de Astillano, para servir y festejar a SS.MM, y entre los cuales figura una libranza, fechada el 13 de enero de 1672, para el pago de diversas mercaderías destinadas al actor Juan Rana (F, XXIX, 34, 105, 115-16); en este año —y no en 1668 (CM5, CLX) o en 1670 (CM9, X, 9) como aventuró Cotarelo— Cosme Pérez, cuando ya estaba retirado, salió en un carro triunfal para actuar en el entremés de <em><strong>El triunfo de Juan Rana</strong></em>, representado junto con la comedia <em><strong>Fieras afemina amor</strong></em>, de Calderón de la Barca, el 29 de enero de 1672 en el Salón Dorado del Buen Retiro. Entre los actores que representaron esa comedia figuran Agustín Manuel, [Antonio de] Escamilla, "actúa en el entremés y en el fin de fiesta", Mariana Romero, María de Quiñones, Antonia del Pozo, "actúa en el fin de fiesta", Josefa López, "actúa en el fin de fiesta", Sebastiana [Fernández], "hace el papel de Calíope y actúa en el fin de fiesta", Micaela [Fernández], "actúa en el fin de fiesta", Manuela [de] Escamilla, "hace el papel de bobo, se viste de hombre y actúa en el fin de fiesta, Mariana [de] Borja "actúa en el fin de fiesta y canta en el sainete", Bernarda Manuela [o Manuela Bernarda 'la Grifona'], "haciendo el papel de Venus", Manuela de Bustamante, Jacinta —actriz que no hemos podido identificar—, "actúa en el fin de fiesta", y Alonso de Olmedo, "hace el papel de Hércules" (F, XXIX, 34, 37, 39); Cosme Pérez murió el 20 de abril en la "calle de Cantarranas, casas propias", según se hacía constar en su partida de defunción, donde se indica que "Recibió los Santos Sacramentos […] Deja 3.400 misas". Fue enterrado en el Convento de Trinitarios de Madrid, y al margen de su partida de defunción, con fecha del 29 de noviembre, se anotaba que se había cumplido su voluntad en lo tocante a las misas por su alma (CM5, CLXI).<br><br><div id="anyo-1673"><b>1673</b><br> </div>Según la <em><strong>Genealogía</strong></em>, su muerte se hacía constar en la Carta de Difuntos de la Cofradía de Nuestra Señora de la Novena de este año, donde se anotaba que murió en Madrid (G, 117).<br><br>
"""
```

```python
# ...existing code...
# </VSCode.Cell>
# <VSCode.Cell id="2f702c54" language="python">
import re
import html
import collections
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# --- 1. PREPROCESAMIENTO DEL TEXTO ---

# Descargar recursos necesarios de NLTK (solo se necesita la primera vez)
try:
    stopwords.words('spanish')
except LookupError:
    print("Descargando stopwords de NLTK...")
    nltk.download('stopwords')
    nltk.download('punkt')

def clean_text(html_content):
    """Extrae, limpia y tokeniza el texto del HTML."""
    # Extraer texto plano del HTML
    text = re.sub(r'<[^>]+>', '', html_content)
    text = html.unescape(text)
    
    # Convertir a minúsculas
    text = text.lower()
    
    # Eliminar números y puntuación
    text = re.sub(r'[\d\W_]+', ' ', text)
    
    # Tokenizar (dividir el texto en palabras)
    tokens = nltk.word_tokenize(text, language='spanish')
    
    # Eliminar stopwords (palabras comunes como 'de', 'la', 'y', etc.)
    stop_words = set(stopwords.words('spanish'))
    # Añadir palabras específicas del contexto que no aportan significado al análisis
    custom_stopwords = {'q', 'v', 'op', 'cit', 'pp', 'cm', 'cl', 'clx', 'clix', 'clviii', 'rs'}
    stop_words.update(custom_stopwords)
    
    cleaned_tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    
    return cleaned_tokens, text

# Obtener el texto del notebook
cleaned_tokens, raw_text = clean_text(html_value)
full_cleaned_text = ' '.join(cleaned_tokens)

# --- 2. ANÁLISIS ESTADÍSTICO BÁSICO ---

total_words = len(raw_text.split())
unique_words = len(set(cleaned_tokens))
total_chars = len(raw_text)
lexical_diversity = (unique_words / len(cleaned_tokens)) * 100 if cleaned_tokens else 0

print("--- ANÁLISIS COMPLETO DEL TEXTO ---")
print("\n### 1. Estadísticas Básicas")
print(f"- Número total de palabras (bruto): {total_words}")
print(f"- Número total de caracteres: {total_chars}")
print(f"- Número de palabras significativas (sin stopwords): {len(cleaned_tokens)}")
print(f"- Número de palabras únicas (vocabulario): {unique_words}")
print(f"- Diversidad léxica: {lexical_diversity:.2f}%")

# --- 3. ANÁLISIS DE FRECUENCIA ---

# Frecuencia de palabras individuales (unigramas)
word_counts = collections.Counter(cleaned_tokens)
most_common_words = word_counts.most_common(15)

print("\n### 2. Palabras más frecuentes")
for word, count in most_common_words:
    print(f"- {word}: {count} veces")

# Frecuencia de bigramas (pares de palabras)
bigrams = nltk.bigrams(cleaned_tokens)
bigram_counts = collections.Counter(bigrams)
most_common_bigrams = bigram_counts.most_common(10)

print("\n### 3. Frases (bigramas) más frecuentes")
for bigram, count in most_common_bigrams:
    print(f"- {' '.join(bigram)}: {count} veces")

# --- 4. VISUALIZACIONES ---

# Gráfico de las palabras más comunes
words, counts = zip(*most_common_words)
plt.figure(figsize=(12, 6))
plt.bar(words, counts, color='skyblue')
plt.title('Top 15 Palabras Más Frecuentes')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Nube de palabras
print("\n### 4. Nube de Palabras")
print("Generando nube de palabras para visualizar los términos clave...")

wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='white',
    colormap='viridis'
).generate(full_cleaned_text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

print("\n--- Fin del Análisis ---")
#
```
