```python
from extractor_relaciones import extraer_relaciones, relaciones_a_csv, mostrar_relaciones
import pandas as pd
import ast
```

```python
# Configuración inicial
years = [1636, 1643, 1648, 1651, 1653, 1656, 1665, 1666, 1670]

# Cargar datos originales
df_original = pd.read_csv("./DicatCSVJuanRana_cleaned.csv")
df_original["YEAR"] = pd.to_numeric(df_original["YEAR"], errors="coerce")

# Cargar datos clarificados
df_clarified = pd.read_csv("./DicatJuanRana_w_clarified_sentences.csv", 
                          sep=";", 
                          encoding="utf-8", 
                          converters={"sentences": ast.literal_eval, 
                                    "clarified_sentences": ast.literal_eval})

print(f"Datos cargados. Procesando {len(years)} años: {years}")

# Cargar datos clarificados
df_clarified2 = pd.read_csv("./DicatJuanRana_w_clarified_sentences_2.0.csv", 
                          sep=";", 
                          encoding="utf-8", 
                          converters={"clarified_sentences": ast.literal_eval})

print(f"Datos cargados. Procesando {len(years)} años: {years}")
```

```python
# Procesar cada año individualmente
resultados_por_año = {}

for año in years:
    print(f"\n{'='*80}")
    print(f"PROCESANDO AÑO: {año}")
    print(f"{'='*80}")
    
    # Filtrar datos para el año actual
    df_año_original = df_original[df_original["YEAR"] == año]
    df_año_clarified = df_clarified[df_clarified["year"] == año]
    df_año_clarified2 = df_clarified2[df_clarified2["year"] == año]
    
    # Extraer textos
    textos_originales = df_año_original["JUAN_RANA_cleaned"].tolist()
    textos_clarified_raw = df_año_clarified["clarified_sentences"].tolist()
    textos_clarified = [" ".join(lista) for lista in textos_clarified_raw]
    textos_clarified_raw2 = df_año_clarified2["clarified_sentences"].tolist()
    textos_clarified2 = [" ".join(lista) for lista in textos_clarified_raw2]
    
    print(f"Textos originales encontrados: {len(textos_originales)}")
    print(f"Textos clarificados encontrados: {len(textos_clarified)}")
    print(f"Textos clarificados 2 encontrados: {len(textos_clarified2)}")
    
    # Inicializar diccionario para este año
    resultados_por_año[año] = {}
    
    # 1. SPACY - TEXTOS ORIGINALES
    if textos_originales:
        print(f"\n{'-'*50}")
        print(f"1. SPACY - TEXTOS ORIGINALES (Año {año})")
        print(f"{'-'*50}")
        
        relacionesSpacy = extraer_relaciones(textos_originales, libreria="spacy", debug=True)
        resultados_por_año[año]['spacy_original'] = relacionesSpacy
        print(f"Se encontraron {len(relacionesSpacy)} relaciones.")
        mostrar_relaciones(relacionesSpacy)
    
    # 2. STANZA - TEXTOS ORIGINALES
    if textos_originales:
        print(f"\n{'-'*50}")
        print(f"2. STANZA - TEXTOS ORIGINALES (Año {año})")
        print(f"{'-'*50}")
        
        relacionesStanza = extraer_relaciones(textos_originales, libreria="stanza", debug=True)
        resultados_por_año[año]['stanza_original'] = relacionesStanza
        print(f"Se encontraron {len(relacionesStanza)} relaciones.")
        mostrar_relaciones(relacionesStanza)
    
    # 3. SPACY - TEXTOS CLARIFICADOS
    if textos_clarified:
        print(f"\n{'-'*50}")
        print(f"3. SPACY - TEXTOS CLARIFICADOS (Año {año})")
        print(f"{'-'*50}")
        
        relacionesSpacyClarified = extraer_relaciones(textos_clarified, libreria="spacy", debug=True)
        resultados_por_año[año]['spacy_clarified'] = relacionesSpacyClarified
        print(f"Se encontraron {len(relacionesSpacyClarified)} relaciones.")
        mostrar_relaciones(relacionesSpacyClarified)
    
    # 4. STANZA - TEXTOS CLARIFICADOS
    if textos_clarified:
        print(f"\n{'-'*50}")
        print(f"4. STANZA - TEXTOS CLARIFICADOS (Año {año})")
        print(f"{'-'*50}")
        
        relacionesStanzaClarified = extraer_relaciones(textos_clarified, libreria="stanza", debug=True)
        resultados_por_año[año]['stanza_clarified'] = relacionesStanzaClarified
        print(f"Se encontraron {len(relacionesStanzaClarified)} relaciones.")
        mostrar_relaciones(relacionesStanzaClarified)
        
    # 5. SPACY - TEXTOS CLARIFICADOS
    if textos_clarified2:
        print(f"\n{'-'*50}")
        print(f"3. SPACY - TEXTOS CLARIFICADOS 2 (Año {año})")
        print(f"{'-'*50}")
        
        relacionesSpacyClarified2 = extraer_relaciones(textos_clarified2, libreria="spacy", debug=True)
        resultados_por_año[año]['spacy_clarified2'] = relacionesSpacyClarified2
        print(f"Se encontraron {len(relacionesSpacyClarified2)} relaciones.")
        mostrar_relaciones(relacionesSpacyClarified2)
    
    # 6. STANZA - TEXTOS CLARIFICADOS
    if textos_clarified2:
        print(f"\n{'-'*50}")
        print(f"4. STANZA - TEXTOS CLARIFICADOS 2 (Año {año})")
        print(f"{'-'*50}")
        
        relacionesStanzaClarified2 = extraer_relaciones(textos_clarified2, libreria="stanza", debug=True)
        resultados_por_año[año]['stanza_clarified2'] = relacionesStanzaClarified2
        print(f"Se encontraron {len(relacionesStanzaClarified2)} relaciones.")
        mostrar_relaciones(relacionesStanzaClarified2)
    
    print(f"\n✓ Año {año} completado.")

print(f"\n{'='*80}")
print("PROCESAMIENTO COMPLETADO")
print(f"{'='*80}")
```

```python
# Mostrar resumen de resultados
print("\nRESUMEN DE RESULTADOS POR AÑO:")
print("="*80)

for año in years:
    if año in resultados_por_año:
        print(f"\nAño {año}:")
        resultados = resultados_por_año[año]
        
        if 'spacy_original' in resultados:
            print(f"  - SpaCy Original: {len(resultados['spacy_original'])} relaciones")
        if 'stanza_original' in resultados:
            print(f"  - Stanza Original: {len(resultados['stanza_original'])} relaciones")
        if 'spacy_clarified' in resultados:
            print(f"  - SpaCy Clarificado: {len(resultados['spacy_clarified'])} relaciones")
        if 'stanza_clarified' in resultados:
            print(f"  - Stanza Clarificado: {len(resultados['stanza_clarified'])} relaciones")
        if 'spacy_clarified2' in resultados:
            print(f"  - SpaCy Clarificado 2: {len(resultados['spacy_clarified2'])} relaciones")
        if 'stanza_clarified2' in resultados:
            print(f"  - Stanza Clarificado 2: {len(resultados['stanza_clarified2'])} relaciones")
    else:
        print(f"\nAño {año}: Sin datos")

print(f"\nTotal de años procesados: {len(resultados_por_año)}")
```
