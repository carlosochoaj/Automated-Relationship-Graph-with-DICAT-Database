```python
from extractor_relaciones import extraer_relaciones, relaciones_a_csv, mostrar_relaciones
import pandas as pd
import ast
```

```python
df = pd.read_csv("./DicatCSVJuanRana_cleaned.csv")

textos = df["JUAN_RANA_cleaned"].tolist()
```

```python
df = pd.read_csv("./DicatJuanRana_w_clarified_sentences.csv", sep=";", encoding="utf-8", converters={"sentences": ast.literal_eval, "clarified_sentences": ast.literal_eval})

textosClarified = df["clarified_sentences"].tolist()
textosClarified = [" ".join(lista) for lista in textosClarified]
```

```python
# Spacy
%%time
relacionesSpacy = extraer_relaciones(textos, libreria="spacy", debug=True)
print(f"\nSe encontraron {len(relacionesSpacy)} relaciones.")
```

```python
mostrar_relaciones(relacionesSpacy)
```

```python
relaciones_a_csv(relacionesSpacy, "relaciones_spacy.csv")
```

```python
# Stanza
%%time
relacionesStanza = extraer_relaciones(textos, libreria="stanza", debug=True)
print(f"\nSe encontraron {len(relacionesStanza)} relaciones.")
```

```python
mostrar_relaciones(relacionesStanza)
```

```python
relaciones_a_csv(relacionesStanza, "relaciones_stanza.csv")
```
