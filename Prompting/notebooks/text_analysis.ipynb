```python
import rich as rc
rc.print(data['cleaned_text'][0])
```

```python
def clean_text(text):
    """
    Function to clean the text data.
    """
    
    text = text.replace('***', '').replace('<br>', '')
    return text
```

```python
data['cleaned_text'] = data['cleaned_text'].apply(clean_text)
```

```python
rc.print(data['cleaned_text'][0])
```

```python
with open('../data/introduccion.txt', 'w', encoding='utf-8') as f:
    for text in data['cleaned_text']:
        f.write(text + '\n')
        break
```

```python
with open('../data/historia_juan_rana.txt', 'w', encoding='utf-8') as f:
    for text in data['cleaned_text'][1:]:
        f.write(text + '\n\n')
```
